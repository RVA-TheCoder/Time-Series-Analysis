```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('EnergyProduction.csv' , index_col = 'DATE' , parse_dates = True)
df.head()
```

```python
df = df.dropna()
```

```python
### Q1 :  setting-up freq

df.index.freq = 'MS'
```

```python
df.index
```

```python
### Q2 : Plotting the dataset

df['EnergyIndex'].plot(figsize = (8,4))
plt.show()
```

### Q3 :  Add a column that shows a 12-month Simple Moving Average (SMA).
###          Plot the result.

```python
df.head()
```

```python
df['SMA_12'] = df['EnergyIndex'].rolling(window = 12).mean()
df.head()
```

```python
df.plot(figsize = (8,4))
plt.show()
```

#### Q4 : Add a column that shows an Exponentially Weighted Moving Average (EWMA) with a span of 12 
####               using the statsmodels SimpleExpSmoothing function. 
####         Plot the result.

```python
from statsmodels.tsa.holtwinters import SimpleExpSmoothing
```

```python
span = 12
alpha = 2/(span +1)
```

```python
model = SimpleExpSmoothing(df['EnergyIndex'])
fitted_model = model.fit(smoothing_level = alpha ,optimized = False )
df['SES_12'] = fitted_model.fittedvalues.shift(-1)
```

```python
df.head()
```

```python
df.plot(figsize = (8,4))
plt.show()
```

#### Q5 : Add a column to the DataFrame that shows a Holt-Winters fitted model using 
####      Triple Exponential Smoothing with multiplicative models. Plot the result.

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
model = ExponentialSmoothing(df['EnergyIndex'] ,
                             trend = 'mul',
                             seasonal = 'mul',
                            seasonal_periods = 12)
fitted_model = model.fit()
df['TES_12'] = fitted_model.fittedvalues
```

```python
df.head()
```

```python
df.plot(figsize = (8,4))
plt.show()
```
